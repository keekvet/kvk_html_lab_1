## Завдання 1(без врахування сили гравітації)

```python
import matplotlib.pyplot as mp
import numpy as np
import math as m

v = 975
angle = m.radians(20)

time_s = 100
iter = 100

arr_t = np.array([t / iter for t in range(0, time_s * iter)])

arr_y = np.array([ v * t * m.sin(angle) for t in arr_t])
arr_x = np.array([ v * t * m.cos(angle) for t in arr_t])

mp.plot(arr_x, arr_y)
```

# Завдання 2 (з гравітацією)

```python
arr_y = np.array([ (v * t * m.sin(angle)) - (9.8 * m.pow(t, 2) / 2) for t in arr_t])
arr_x = np.array([ v * t * m.cos(angle) for t in arr_t])

plt = mp.plot(arr_x, arr_y)
axes = mp.gca()
axes.set_ylim(0)
```
